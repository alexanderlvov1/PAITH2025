### Plotting in Matplotlib can be done in either of two ways, which ones? Which way is the recommended approach?

```python
#1) Object-Oriented (OO) approach - Creating explicit Figure and Axes objects
#2) Pyplot (state-based) approach - Using plt.plot(), plt.title(), etc. directly
# The OO is the reomemmended but both are accaptable.
```

### Explain shortly what a figure, axes, axis and an artist is in Matplotlib.

```python
# Figure is the whole window or page that contains everything. 
# Axes is a single plot area withing a figure where data is plotted. It has x and y axes. One figure can contain multiple axes or subplots.
# Axis is the invidual lines (x-axis or y-axis) that the define the scale and liits the of the plot. Each Axes typically has two axis (x and y).
# Artist is everything one can see on a figure. This includes the figure itself, area, axis, lines, text, labels etc.
```
